```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol
ticker = yf.Ticker("OJ=F")

# Fetch historical data from January 1, 2002 to December 31, 2024
df = ticker.history(start="2002-01-01", end="2025-01-01")

# Reset the index to have 'Date' as a column
df.reset_index(inplace=True)

# Create a new DataFrame with only the 'Date' and 'Close' columns
df_filtered = df[['Date', 'Close']]

# Convert 'Date' to timezone-naive, then to the 'YYYY-MM-DD' format
df_filtered.loc[:, 'Date'] = df_filtered['Date'].dt.tz_localize(None).dt.strftime('%Y-%m-%d')

# Rename the columns
df_filtered.columns = ['DATE', 'CLOSE PRICE']

# Save the new DataFrame to a CSV file
df_filtered.to_csv('orange_juice_close_prices.csv', index=False)

# Display the first few rows of the new DataFrame to confirm
print(df_filtered.head())
```
